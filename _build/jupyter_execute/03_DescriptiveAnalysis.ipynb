# Deskriptive Analyse von Daten

Jede gute Datenanalyse beinhaltet eine deskriptive Untersuchung der Daten. Dies erfolgt meist eher zu Beginn einer Auswertung, um sich einen ersten Eindruck zu den Merkmalsausprägungen der Daten und potentiellen Zusammenhängen zu machen. Zur deskriptiven Analyse gehört die Interpretation der Verteilung einzelner Merkmale und auch deren Zusammenhänge. Passend für diesen Einsatzbereich sind quantitative Kennzahlen aus der beschreibenden Statistik und graphische Darstellungen. Wir wollen uns in diesem Kapitel einen kleinen Überblick über gängige Kennzahlen und Graphiken verschaffen und diskutieren, welche Schlüsse hieraus gezogen werden können. 

## Datenerhebung und Datenkollektion

Jeder Datenanalyse geht die Erhebung von Daten voraus. Ziel der Analyse ist es möglichst allgemein gültige Zusammenhänge zu identifizieren. Beispielsweise könnten wir uns dafür interessieren, welche Faktoren den Mietpreis von Mietwohungen in Deutschland beeinflussen. Die relevante Einheit unseres Beispiels wären die Wohnungen. Wären wir in der Lage Daten zu allen Wohnungen zu sammeln, so hätten wir den Überblick über die Grundgesamtheit. Dies ist bereits in diesem Beispiel mit offensichtlich hohem Aufwand verbunden bzw. sogar unmmöglich, wenn einige der Immobilieninhaber*innen nicht gewillt sind, die Daten mit uns zu teilen. Zudem stellt sich die Frage, welche Merkmale neben dem Mietpreis gemessen werden sollten, die potentiell Einfluss auf diesen haben. 

Das einzige was als Lösung bleibt, ist es in der Regel eine Stichprobe der Einheiten (hier Wohnungen) zu ziehen und hierbei möglichst viel sinnvolle Merkmalsausprägungen je Einheit abzufragen. Beziehen wir die Stichprobe bereits von Dritten, müssen wir in der Regel mit dem was zugänglich ist, die Analyse durchführen und zu einem späteren Zeitpunkt entscheiden, ob wir die verschiedenen Merkmale als ausreichend für den Erkenntnisgewinn der Ananlyse betrachten.

Eine Stichprobe sollte repräsentativ und zufällig, jedoch nicht willkürlich gewählt werden. In den meisten Fällen macht es am meisten Sinn, wenn jede Einheit mit der gleichen Wahrscheinlichkeit für eine Stichprobe gezogen wird. Ohne näher auf die statistischen Hintergründe einzugehen, kann man festhalten, dass die Erkenntnisse aus den Analysen der Stichprobe nur dann Rückschlüsse auf allgemeine Zusammenhänge zulassen, wenn die Stichprobe in etwa die Zusammenhänge der Daten der Grundgesamtheit widerspiegelt. Dennoch treten leichte unterschiedliche Resultate der Stichprobe bedingt durch die Zufälligkeit der Stichprobenziehung und der damit verbundenen Unvollständigkeit gegenüber der Gesamtpopulation mit Sicherheit auf. Die Quantifizierung dieser Unterschiedlichkeit ist ein zentraler Bestandteil der Statistik, jedoch nicht im zentralen Fokus dieses Kurses. Dennoch sollen die genannten Argumente für einen vorsichtigen Umgang mit den Ergebnissen der Datenanalyse sensibilisieren.

## Umgang mit fehlenden Daten

In vielen Datensätzen fehlen oft Werte, dies ist meist durch Kürzel wie "NA, NAN, na, nan" gekennzeichnet. Wir wollen es im Umgang mit fehlenden Werten in diesem Kurs einfach halten und fokussieren uns daher auf die eher einfachen bestehenden Lösungen. Die beiden einfachsten Möglichkeiten bestehen entweder darin die Zeile (also eine Beobachtung was in der Regel den Ausprägungen einer statistischen Einheit entspricht) zu löschen, in der es zu mindestens einem fehlenden Wert kommt. Dies macht Sinn, wenn der gesamte Datensatz dadurch nicht zu stark reduziert wird. Kommt es vor allem bei einem Merkmal zu häufigen fehlenden Werten, kann auch überlegt werden, nur diese Spalte, also alle Ausprägungen eines Merkmals zu löschen. Alternativ können fehlende Werte durch möglichst repräsentative Werte ersetzt werden. Dies sollte jedoch nicht zu leichtfertig gemacht werden, da auf diese Art eine Verzerrung der Realität bei der Analyse eingeführt wird. Dennoch kann es die bessere Lösung sein, wenn auf anderem Weg die Datengrundlage zu stark eingeschränkt wird. 

## Merkmalstypen und Skalen

Oft sind Datensätze eine Sammlung von Merkmalsausprägungen verschiedener Variablen von Einheiten. Man unterscheidet bei Merkmalen ob diese diskret (mit endlich abzählbar oder unendlich abzählbar vielen möglichen Realisierungen) oder stetig (alle Realisierungen in einem Intervall sind möglich) sind. Zudem ist das Skalenniveu eines Merkmals wichtig für die Datenanalyse. Folgende Skalenniveus werden unterschieden:

* nominal: Kategorien, ohne sinnvolle Reihung
* ordinal: mögliche Reihung, keine sinnvolle Interpretation der Abstände
* kardinal bzw. metrisch: mögliche Reihung und Abstandsinterpretation, zum Teil jedoch ohne sinnvollen Nullpunkt (Intervallskala vs. Verhältnisskala)

In der unteren Zelle sehen wir beispielsweise einen Datensatz von Schülern, bei dem es um eine mögliche Analyse zwischen den Zusammenhang zwischen der Prüfungsleistung (G3) und weiteren Merkmalen geht. Die Variable "sex" bezieht sich z.B. auf das Geschlecht und ist kategorial, hier ist es durch "F" für female und "M" wie male abgekürzt. Um die Ausprägungen dieser Variable später analysieren zu können bietet es sich an die Kategorien in Zahlenwerte zu transformieren. Dennoch machen auch nach der Kodierung mathematische Berechnungen wie die Addition inhaltlich keinen Sinn. "Fedu" und "Medu" sind gemäß der Variablenbeschreibung klar ordinal. Beispielsweise das Alter "age" ist metrisch. 

```python
import pandas as pd


# import original data
df_math = pd.read_csv("../Daten/student/student-mat.csv", sep = ";")

# variable descriptions
# sex - student's sex (binary: "F" - female or "M" - male)
# age - student's age (numeric: from 15 to 22)
# Medu - mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
# Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)
# studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
# freetime - free time after school (numeric: from 1 - very low to 5 - very high)
# absences - number of school absences (numeric: from 0 to 93)
# G3 - final grade (numeric: from 0 to 20, output target)
df_math = df_math.loc[:, ["sex", "age", "Medu", "Fedu", "studytime", "freetime", "absences", "G3"]]
df_math.head()
```

```python
# numerical transformation of the variable sex
df_math.loc[:, "num_sex"] = df_math.sex.replace({"F": 1, "M": 0})
df_math.astype({"num_sex": "int32"})
df_math.drop(["sex"], axis = 1, inplace = True)
df_math.head()
```

## Quantitative Beschreibung der Merkmale

Um einen ersten Überblick über die einzelnen Merkmalsausprägungen zu erhalten, bieten es sich an die Verteilungen dieser näher zu betrachten. Dies kann anhand von beschreibenden Kennzahlen aus der deskriptiven Statistik und anhand von geeigneten Grafiken durchgeführt werden. Gängige Kennzahlen sind das arithmetische Mittel, der Median, Minima, Maxima, Quantile, die Standardabweichung, Schiefe und Wölbung. Zur Wiederholung ein paar Definitionen.

Für eine Anzahl von $n$ Beobachtungen eines Merkmals $x_1, ..., x_n$ ist das arithmetische Mittel definiert durch:

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i
$$

Gegeben eine geordnete Liste der Merkmalsausprägungen $x_{(1)}, ..., x_{(n)}$, ist der Median definiert durch:

$$
x_{med} = 
\begin{cases}
 x_{\left( \frac{n+1}{2} \right)} & \text{wenn } n \text{ ungerade}\\
 \frac{1}{2} \left(x_{(\frac{n}{2})} + x_{(\frac{n}{2}+1)} \right) & \text{wenn } n \text{ gerade}
\end{cases}
$$

Das $p$ Quantil eines Merkmals ist der Wert für den ein Anteil der Ausprägungen $p$ kleiner oder gleich und mindestens ein Anteil $1-p$ größer oder gleich ist. Minimum und Maximum ist die jeweils kleinste bzw. größte aller Ausprägungen. Um die Schwankung der Ausprägungen untersuchen zu können, wird meist die empirische Standardabweichung verwendet. Die Varianz ist die mittlere quadratische Abweichung der Werte vom Mittelwert.

$$
s^2 = \frac{1}{n} \sum_{i=1}^n \left( x_i - \bar{x} \right)^2
$$

Die empirische Standardabweichung ist die Wurzel der empirischen Varianz, $s = \sqrt{s^2}$. Um die Schiefe der Verteilung empirisch zu analysieren wird oft der Momentenkoeffizient der Schiefe verwendet:

$$
g_m = \frac{\frac{1}{n} \sum_{i - 1}^n \left( x_i - \bar{x} \right)^3}{s^3}
$$

Ein negativer Wert spricht für eine linksschiefe, ein positiver Wert für eine rechtsschiefe Verteilung. Bei einem Wert von $0$ oder nahe $0$ kann man von einer symmetrischen Verteilung ausgehen. 

Die Wölbung der Verteilung wird oft mit dem Wölbunsmaß nach Fisher beurteilt:

$$
\gamma = \frac{\frac{1}{n} \sum_{i - 1}^n \left( x_i - \bar{x} \right)^4}{s^4} - 3
$$

Beachten Sie, dass beim letzten Maß der Wert von $3$ abgezogen wird, um die Wölbung mit der einer Normalverteilung zu vergleichen. Bei einem Wert größer $0$ spricht man von einer leptokurtischen, bei einem Wert kleiner $0$ von einer platokurtischen Verteilung. Manche Statistikprogramme ziehen nicht per default den Wert von $3$, dies gilt es zur Sicherheit meist noch zu überprüfen.

Wenn wir mit pandas DataFrames arbeiten, ist es relativ einfach diese deskriptiven Maße zu bestimmen. Zum einen besteht die *describe* Funktion, wodurch Anzahl der Beobachtungen, Mittelwert, Standardabweichung, Minimum, Maximum, Median sowie das 25\% und 75\% Quantil ausgegeben werden. Darüber hinaus existieren die *skew* und die *kurtosis* Funktion.

## Graphische Darstellung der Daten

Es existieren viele anschauliche Möglichkeiten Daten zu visualisieren. Um möglichst erkenntnisreiche Informationen aus den Visualisierungen zu ziehen bedarf es jedoch Erfahrung und Fachkentniss. Wir wollen daher zunächst mit ein paar grundlegenden Grafiktypen beginnen. Die Wahl der passenden Grafik ist auch hier wieder in Teilen mit der Skala eines Merkmals verbunden. Um einen Überblick über die Verteilung von Merkmalsausprägungen zu erhalten bieten sich für nominal oder ordinal skalierte Merkmale Barplots (Stabdiagramme) und für metrisch skalierte Merkmale Histogramme an. 


### Univariate Verteilungen der Merkmalsausprägungen

In der unteren Zelle sehen wir Barplots für die kategorialen bzw. ordinalen Merkmale. Bei den Barplots werden auf der x-Achse die Merkmalsausprägungen und auf der y-Achse die absoluten Häufigkeiten abgetragen. Wir können anhand dieser Darstellungen erkennen, dass die meisten Verteilungen der Merkmalsausprägungen nicht symmetrisch sind und insbesondere bei der "studytime" und "freetime" meist eine Kategorie deutlich häufiger als alle anderen Kategorien bei den Ausprägungen auftritt. 

```python
import matplotlib.pylab as plt


non_metric_vars = ["Medu", "Fedu", "studytime", "freetime", "num_sex"]

fig_non_metric, axs_non_metric = plt.subplots(1, 5, figsize = (14, 5))
for i, non_m_var in enumerate(non_metric_vars):
    nobs_per_cat = df_math.groupby(non_m_var).size()
    axs_non_metric[i].bar(nobs_per_cat.index, nobs_per_cat.values)
    axs_non_metric[i].set_title(non_m_var)
plt.show()
```

In der nächsten Zelle sehen wir die Histogramme der metrischen Variablen. Bei einem Histogramm werden die Häufigkeiten der Ausprägungen innerhalb definierter Intervalle gezählt. An Stelle fester Intervallgrenzen wird auch oft eine Anzahl von Internallen (bins) vorgegeben, und die Intervallgrenzen werden entsprchend über die Bandbreite aller Merkmalsausprägungen bestimmt. Alternativ zur absoluten Anzahl der Merkmalsausprägungen innerhalb eines Intervalls, werden auf der y-Achse auch manchmal die empirischen Dichtewerte abgetragen. Hierbei werden die Häufigkeiten so normalsiert, dass die Fläche unterhalb des Histograms den Wert $1$ ergibt. Letzteres wird vor allem dann vorgezogen, wenn die empirische Verteilung mit Dichtefunktionen theoretischer Verteilungen verglichen werden soll.

```python
import matplotlib.pylab as plt


metric_vars = ["age", "absences", "G3"]

fig_metric, axs_metric = plt.subplots(1, 3, figsize = (12, 5))
for i, m_var in enumerate(metric_vars):
    axs_metric[i].hist(df_math[m_var], bins = 10, density = True)
    axs_metric[i].set_title(m_var)
plt.show()
```

Eine weitere Möglichkeit eine graphischen Eindruck der Verteilungen zu erhalten, besteht durch die Verwendung von Boxplots, welche sowohl für nicht-metrische wie auch für metrische Variablen eingesetzt werden können. Im Vergleich zu den Barplots und den Histogrammen sind Boxplots besser geeignet, um einen Eindruch von extremen Werten und deren Häufung zu erhalten. Die nächste Zelle visualisiert die Boxplots aller Merkmale im Datensatz. Im Standard, der in der unteren Zelle gezeigt wird, entspricht die untere Kante der Box dem 25\% und die obere Kante der Box dem 75\% Quantil, der Strich in der Mitte symbolisiert den Median. Die äußeren Striche sind Abstände in Höhe des $1,5$-fachen des Interquartilsabstands. Als extremeren Ausprägungen werden durch einzelne Punkte dargestellt. Mit Hilfe der Boxplots im Beispiel lässt sich beispielsweise gut erkennen, dass es bei der Abwesenheit deutlich häufiger zu vereinzelt hohen Werten kommt, als bei den übrigen Merkmalen.

```python
boxplots = df_math.boxplot(figsize = (12, 6))
plt.show()
```

### Darstellung von Zusammenhängen

Sehr oft sind Zusammenhänge zwischen den Merkmalsausprägungen einzelner Variablen ebenso von großer Bedeutung für die Datenanalyse. Im ersten Schritt bieten sich Scatterplots an, um paarweise Zusammenhänge zwischen Variablen zu visualisieren. Diese liefern jedoch meist nur Erkenntnisse im Falle von stetigen Variablen. Beispielsweise sehen wir in der unteren Zelle einen Scatterplot zwischen der Abwesenheit und der finalen Note. Hieraus schein durchaus ein Zusammenhang ersichtlich zu sein. Daneben ist ein Scatterplot zwischen der Studien- und der Freizeit. Da beide Variablen in diesem Datensatz diskret sind, kann man aus dieser Ansicht nur erkennen, welche Werte-Kombinationen auftreten, was jedoch wenig erkenntnisreich ist. Als kleinen "Workaround" wäre es für solche Fälle möglich, die Größe der Punkte der Grafik proportional zur Häufigkeit der paarweisen Kombinationen darzustellen. Diese Graphik können Sie im dritten Plot betrachten. 

```python
import matplotlib.pylab as plt

fig, axs = plt.subplots(1, 3, figsize = (12, 5))

df_math.plot.scatter(x = "absences", y = "G3", ax = axs[0])
df_math.plot.scatter(x = "freetime", y = "studytime", ax = axs[1])

count_study_free = df_math.groupby(["freetime", "studytime"]).size()
x_levels = count_study_free.index.get_level_values(level = 0)
y_levels = count_study_free.index.get_level_values(level = 1)

for x in x_levels:
    for y in y_levels:
        axs[2].scatter(x, y, s = 2 * count_study_free.loc[x, y], color = "#1f77b4")

axs[2].set_xlabel("freetime")
axs[2].set_ylabel("studytime")

axs[1].set_title("ohne Häufigkeitsinformation")
axs[2].set_title("mit Häufigkeitsinformation")

plt.show()
```

Die Möglichkeit alle paarweisen Abhängigkeiten eines Datensatzes zu visualisieren, besteht durch die Verwendung von Heatmaps der Korrelationsmatrix eines Datensatzes. Auch hier gilt es bei der Berechnung der Korrelationen, die Skalen der Merkmale zu beachten. Solange ein Merkmal mindestens ordinal skaliert ist, kann der Korrelationskoeffizient nach Spearman (Rangkorrelation) verwendet werden. In unserem Beispiel schließen wir daher die Variable des Geschlechts aus. In einer Heatmap werden durch Farben und Zahlen relativ schnell alle paarweisen Abhängigkeiten übersichtilich. Ein Beispiel finden Sie in der folgenden Zelle.

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt


corr_mat = df_math.drop(["num_sex"], axis = 1).corr(method = "spearman")
mask = np.triu(np.ones_like(corr_mat))

fig, ax = plt.subplots()
sns.heatmap(corr_mat, annot= np.round(corr_mat,2), mask=mask, cmap = "coolwarm", ax = ax)
plt.show()
```

Zuletzt betrachten wir noch ein Beispiel für die Zusammenhangsanalyse, wenn wir an einer Zielvariable und den möglichen Einfluss von einzelnen unabhängigen Merkmalen interessiert sind, die nicht stetig skaliert sind. Es bietet sich hier an, Histogramme nach den Kategorien der unabhängigen Variablen zu generieren. Beispielsweise sehen wir in der Graphik der nächsten Zelle erneut in Histogrammform der finalen Prüfungsergebnisse, jedoch dieses Mal getrennt nach Personen mit hoher und geringerer Lernzeit. Derartige Grafiken lassen sich auch für Merkmale mit mehreren Ausprägungen erstellen. Gleichen sich die Histogramme des Merkmals von Interesse für unterschiedliche Kategorien eines anderen Merkmals, kann eher weniger von einem Einfluss des Merkmals auf die Zielvariable ausgegangen werden. Je unterschiedlicher die Verteilungen in Abhängigkeit der Kategorie, umso stärker ist potentiel der Einfluss des Merkmals.

```python
import matplotlib.pylab as plt


fig, ax = plt.subplots(figsize = (8, 5))

ax.hist(df_math["G3"][df_math.studytime <= 2], alpha = 0.7, label = "studytime <= 2")
ax.hist(df_math["G3"][df_math.studytime > 2], alpha = 0.7, label = "studytime > 2")
ax.set_title("Final test results")
ax.legend()
plt.show()
```

Hiermit beenden wir den Auszug aus den Möglichkeiten der graphischen Beschreibung von Daten. An dieser Stelle sei angemerkt, dass die Erzeugung erkenntnisreicher Graphiken ein lebenslanger Lernprozess ist, da gute graphische Visualisierungen stark vom jeweiligen Datensatz abhängen.

## Vorbereitung der Daten

Nach den ersten Erkenntnissen der deskriptiven Analyse, müssen manche Datenpunke vorbereitet werden, um Modelle der Datenanalyse mit diesen Daten zu trainieren. Die relevantesten Aspekte hierbei sind der Umgang mit kategorialen Daten und die Normierung quantitativer Merkmalsausprägungen. Bei nominalen und ordinalen Merkmalen, kann es vorkommen, dass diese im Datensatz als nicht numerische Werte übergebene werden, z.B. "male", "female" oder "bad", "neutral", "good". Für orinale Merkmale bietet es sich an, den Kategorien Zahlen entsprechend der oridinalen Reihung zuzuordnen, z.B. "bad" -> 1, "neutral" -> 2, "good" -> 3. Es ist übrigens nicht zwingend erforderlich, dass die Zahlen 1,2,3 verwendet werden, solange eine sinnvolle Logik bei der Reihung verwendet wird, können die Ausprägungen beliebig gewählt werden. Ist ein Merkmal nominal skaliert, wird häufig eine 1-0 Kodierung verwendet. Im Englischen spricht man vom One-Hot-Encoding dessen Name impliziert, dass die jeweils zutreffende Kategorie als "Hot" durch eine 1 markiert wird. Genauer muss beim One-Hot-Encoding so viele neue Variablen generieren, wie es Kategorien gibt. Bei zwei Kategorien würde beispielsweise die erste Kategorie durch den Vektor $\begin{pmatrix} 1 & 0 \end{pmatrix}^T$ und die zweite Kategorie durch den Vektor $\begin{pmatrix} 0 & 1 \end{pmatrix}^T$, bei drei Kategorien würde beispielsweise die erste Kategorie durch den Vektor $\begin{pmatrix} 1 & 0 & 0 \end{pmatrix}^T$, die zweite Kategorie durch den Vektor $\begin{pmatrix} 0 & 1 & 0 \end{pmatrix}^T$ und die dritte Kategorie durch den Vektor $\begin{pmatrix} 0 & 0 & 1 \end{pmatrix}^T$ dargestellt werden. Diese Art der Kodierung kann auch für ordinale Merkmale verwendet werden, jedoch geht hiermit die ordinale Information zunächst verloren. Alternativ wird in der Statistik auch die Dummy-Kodierung verwendet, bei der eine Kategorie als Standard festgelegt und die übrigen Ausprägungen durch $K-1$ 1-0 Variablen kodiert werden. 

Bei numerischen Merkmalen kann es oft sinnvoll und manchmal auch erforderlich sein, die Werte auf einen vergleichbaren Wertebereich zu bringen. Dies erleichtert bei der Analyse die Vergleichbarkeit der Einflussstärke und ist für manche Algorithmen erforderlich um numerische Probleme beim Trainieren der Modelle zu verhindern. Für diesen Zweck, können unterschiedliche Arten der Normierung gewählt werden. Zwei gängie Möglichkeit sind die Standardisierung:

$$
z = \frac{x - \bar{x}}{s}
$$

und die Min-Max-Skalierung:


$$
z = \frac{x - \min(x)}{\max(x) - \min(x)}
$$

In den unteren Zellen generieren wir einen Beispieldatensatz mit den Merkmalen "gender", "assessment" und "score". Die *info* Methode des pandas-DataFrames zeigt uns, dass es sich bei den ersten beiden Variablen um kategoriale Merkmale handelt, die im Textformat im Datensatz enthalten sind. Dies wird von pandas durch den Typ "object" signalisiert. Der Typ "float64" beschreibt eine floating point number, was eine Kommatrennzahl entspricht womit wir von einem numerischen Wert ausgehen können. Im ersten Schritt wird das Merkmal "gender" auf die Anzahl unterschiedlicher Ausprägungen analysiert und im Anschluss One-Hot-Encoded. Im Anschluss werden den Kategorien der Variable "assessment" in aufsteigender Reihenfolge Zahlen zugewiesen. Zuletzt wird die Variable "score" standardisiert und Min-Max skaliert. Die Auswirkung der unterschiedlichen Skalierung lässt sich am besten durch den Vergleich der Histogramme oder der Boxplots dieser Variablen visualisieren.

```python
import pandas as pd
import numpy as np
import random

gender = ["male"] * 250 + ["female"] * 250
random.shuffle(gender)
assessment = ["bad"] * 150 + ["neutral"] * 200 + ["good"] * 150
random.shuffle(assessment)
numerical = np.random.normal(size = 500, loc = 2, scale = 2)

df = pd.DataFrame({"gender": gender, "assessment": assessment, "score": numerical})
print(df.info())
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder


one_hot_encoder = OneHotEncoder()

print("The variable gender has the following unique values:")
print(df.gender.unique())

df.loc[:, ["female", "male"]] = one_hot_encoder.fit_transform(df.gender.values.reshape(-1, 1)).toarray()
df.drop(["gender"], axis = 1, inplace = True)
df.head()
```

```python
print("The variable assessment has the following unique values:")
print(df.assessment.unique())

df.replace(to_replace = ["bad", "neutral", "good"], value = [1, 2, 3], inplace = True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

df.loc[:, "standard_score"] = standard_scaler.fit_transform(df.score.values.reshape(-1, 1))
df.loc[:, "minmax_score"] = minmax_scaler.fit_transform(df.score.values.reshape(-1, 1))
df.head()
```

```python
import matplotlib.pylab as plt

fig, axs = plt.subplots(1, 3, figsize = (12, 5))
axs[0].hist(df.score)
axs[1].hist(df.standard_score)
axs[2].hist(df.minmax_score)
axs[0].set_title("score")
axs[1].set_title("standard_score")
axs[2].set_title("minmax_score")
plt.show()
```

```python
df.loc[:, ["score", "standard_score", "minmax_score"]].boxplot()
plt.show()
```
